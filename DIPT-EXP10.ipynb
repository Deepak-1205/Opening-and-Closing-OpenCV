```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_img():
    blank_img =np.zeros((500,700))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(blank_img,text='Deepak',org=(50,300), fontFace=font,fontScale= 5,color=(255,255,255),thickness=25,lineType=cv2.LINE_AA)
    return blank_img
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
    plt.show()
```

```python
img = load_img()
```

```python
display_img(img)
```

```python
#Erosion
```

```python
kernel = np.ones((5,5),dtype=np.uint8)
```

```python
kernel
```

```python
erosion1 = cv2.erode(img,kernel,iterations=1)
```

```python
display_img(erosion1)
```

```python
#Opening
```

```python
img = load_img()
```

```python
white_noise = np.random.randint(low=0,high=3,size=(500,700)) 
```

```python
white_noise  # Display
```

```python
white_noise = white_noise*255
```

```python
white_noise # Display 
```

```python
noise_img = white_noise+img
```

```python
display_img(noise_img)   
```

```python
opening = cv2.morphologyEx(noise_img, cv2.MORPH_OPEN, kernel)
```

```python
display_img(opening)
```

```python
#Closing
```

```python
img = load_img()
```

```python
black_noise = np.random.randint(low=0,high=2,size=(500,700))
```

```python
black_noise
```

```python
black_noise= black_noise * -255
```

```python
black_noise_img = img + black_noise
```

```python
black_noise_img     # display
```

```python
black_noise_img[black_noise_img==-255] = 0
```

```python
display_img(black_noise_img)
```

```python
closing = cv2.morphologyEx(black_noise_img, cv2.MORPH_CLOSE, kernel)
```

```python
display_img(closing)
```
